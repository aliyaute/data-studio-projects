```python
def scrape_page(row):
    url = f"https://reconnectingasia.csis.org/database/projects/" + row['url']
    print("Scraping", url)


    response = requests.get(url)
    doc = BeautifulSoup(response.text, 'html.parser')


    page = {}

    try:
        page['name'] = doc.find('h1').string.replace('\n', '').replace(' ', '')
        print("----")
        tags = doc.select(".details p")
        page['project_type'] = tags[0].text
        page['status'] = tags[1].text

        try:
            page['start_date'] = tags[4].text
        except:
            pass
        page['total_reported_cost'] = tags[2].text

    except:
        pass
    # In your function, always return pd.Series(page)
    # It converts what you've scraped into a dataframe
    return pd.Series(page)
```

```python
scraped_df = df.apply(scrape_page, axis=1)
scraped_df
```
